# Hello World! in Tensorflow & Keras

## Imports

Standard imports for Tensorflow, Keras, and NumPy

```python
import tensorflow as tf
import keras
import numpy as np
```

## Defining and Compiling the NN

The neural network has 1 layer, and that layer has 1 neuron. The input shape is one value. To compile, we need to functions, an optimizer and a loss function. Based on what the loss function returns, the optimizer attempts to go closer to the "correct" fit. 

SGD = Stochastic Gradient Descent

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

## Using NumPy to Store the Data

Declaring the array in NumPy to feed in the values. np.array([], dtype)

```python
x = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
y = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

## Fitting the Model Using 500 Epochs

Running the optimizer and loss function to make a closer and closer guess 500 times.

```python
model.fit(x, y, epochs=500)
```

## Prediction

Not exactly 19, we are dealing with probabilities not certainties. Not enough data as well.

```python
print(model.predict([10.0]))
```
